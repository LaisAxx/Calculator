```
# O pacote pandasql (https://pypi.org/project/pandasql/) nos permite fazer 
# consultas em dataframes usando o SQLite syntax.

import pandas
!pip install -U pandasql
import pandasql

def select_first_50(filename):
  
  # Carrega o arquivo aadhaar_data.csv em um dataframe do pandas. Na sequência,
  # renomeia as colunas, substituindo os espaços por underscores e transformando
  # todos os caracteres para minúsculo, de forma que se assemelhe mais a uma 
  # tabela de banco de dados.
  aadhaar_data = pandas.read_csv(filename)
  #ajusta colunas do dataframe 
  collmns = []
  for c in aadhaar_data.columns:
    new_col = c.lower().replace(' ', '_')
    collmns.append(new_col)
  aadhaar_data.columns = collmns  
  

  # Selecione os primeiros 50 registros para as colunas "registrar" e 
  # "enrolment_agency" da tabela aadhaar_data usando a sintaxe SQL. 
  #
  # Você pode fazer o download de uma cópia dos dados aadhaar no link abaixo
  # para solucionar este exercício:
  # https://1drv.ms/u/s!Av8pSmXDlugUjdMAWEIGGqx_6MS7Tg
  # ou
  # https://drive.google.com/file/d/1jEf75RgFwp9EtYCBHPu9eWSD_jZ8wf-D/view

  # Query SQL
  q = """
  select registrar, enrolment_agency 
  from aadhaar_data
  limit 50
  """

  # Executa o seu comando SQL e retorna um dataframe
  aadhaar_solution = pandasql.sqldf(q.lower(), locals())
  return aadhaar_solution    
```

## 3.11 - Escreva a sua própria consulta complexa (enviar pelo Canvas)

```
import pandas
!pip install -U pandasql
import pandasql

def aggregate_query(url):
  # Carrega o arquivo aadhaar_data.csv em um dataframe do pandas. Na sequência,
  # renomeia as colunas, substituindo os espaços por underscores e transformando
  # todos os caracteres para minúsculo, de forma que se assemelhe mais a uma 
  # tabela de banco de dados.
  aadhaar_data = pandas.read_csv(url, encoding='cp1252')
  aadhaar_data.rename(columns = lambda x: x.replace(' ', '_').lower(), inplace=True)
  

  # Escreva uma query que selecione da tabela aadhaar_data a quantidade de
  # homens e mulheres acima de 50 anos que tiveram a identificação aadhaar
  # gerada para cada distrito.
  # aadhaar_generated é a coluna que denota a quantidade de Aadhaar IDs foram
  # gerados.
  #
  # Possíveis colunas para serem selecionadas dos dados de Aadhaar:
  #     1) registrar
  #     2) enrolment_agency
  #     3) state
  #     4) district
  #     5) sub_district
  #     6) pin_code
  #     7) gender
  #     8) age
  #     9) aadhaar_generated
  #     10) enrolment_rejected
  #     11) residents_providing_email,
  #     12) residents_providing_mobile_number
  #
  # Você pode fazer o download dos dados de Aadhaar pelo link abaixo:
  # https://1drv.ms/u/s!Av8pSmXDlugUjdMAWEIGGqx_6MS7Tg
  # ou
  # https://drive.google.com/file/d/1jEf75RgFwp9EtYCBHPu9eWSD_jZ8wf-D/view

  q = """
    select gender, district, sum(aadhaar_generated) from
    aadhaar_data where age > 50
    group by gender, district
  """

  # Executa o seu comando SQL e retorna um dataframe
  aadhaar_solution = pandasql.sqldf(q.lower(), locals())
  return aadhaar_solution    
```

```
# Neste exercício, você deseja chamar a API last.fm para obter uma lista dos 
# melhores artistas da Espanha (método geo.getTopArtists).

# Sua função deve devolver o nome do cantor ou cantora número 1 na Espanha.

# Sua principal tarefa, aqui, é navegar pela estrutura de dados retornada pela 
# API last.fm para encontrar e retornar o melhor cantor da Espanha.

import json
import requests
import time
import pandas as pd

def get_top_artist():

  #Seu código vai aqui
  api_key = ''
  artist_name = []
  artist_listeners = []
  for page in range(1,3):
    # print(f'Página: {page}')
    url= f"http://ws.audioscrobbler.com/2.0/?method=geo.gettopartists&country=spain&api_key={api_key}&format=json"
    data = requests.get(url).text
    data = json.loads(data)
  for artist in data['topartists']['artist']:
    artist_name.append(artist['name'])
    artist_listeners.append(artist['listeners'])
  
  time.sleep(1)

  df = pd.DataFrame({'name':artist_name, 'listeners': artist_listeners})
  #top 5
  # df.sort_values(by='listeners', ascending=False).head()

  max_listeners = df['listeners'].max()
  return df[df['listeners'] == max_listeners]


# if __name__ == '__main__':
#   api_key = ''
#   print(get_top_artist(api_key))
```
